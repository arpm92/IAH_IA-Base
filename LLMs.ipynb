```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load the pre-trained model and tokenizer from Hugging Face
model_name = "gpt2"  # You can also try other models like 'gpt2-large', 'gpt2-xl', or 'EleutherAI/gpt-neo-2.7B'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Move the model to GPU (if available)
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
# Define the prompt for text generation
prompt = "In a world where artificial intelligence is used for everyday tasks,"
```

```python
# Encode the input prompt into tokens
input_ids = tokenizer.encode(prompt, return_tensors='pt').to(device)

# Generate text based on the prompt
output = model.generate(input_ids,
                        max_length=100,  # Maximum length of generated text
                        num_return_sequences=1,  # Number of different sequences to return
                        no_repeat_ngram_size=2,  # Avoid repeating n-grams
                        early_stopping=True,  # Stop generating once a good completion is found
                        do_sample=True,  # Use sampling instead of greedy decoding for diversity
                        top_k=50,  # Top-k sampling
                        top_p=0.95  # Top-p (nucleus) sampling
                       )

# Decode and print the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("\n\n")
print(generated_text)
print("\n\n")
```
